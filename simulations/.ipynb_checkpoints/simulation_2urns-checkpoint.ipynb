```python
from pathlib import Path
from concurrent.futures import ThreadPoolExecutor, as_completed
import numpy as np
import matplotlib.pyplot as plt
from utils import *  
```

## Hyper-parameters

```python
n, xi = 800, 350
delta_grid = np.arange(0.05, 0.51, 0.05)

N_PERM, ALPHA, R = 300, 0.05, 15
BASE_SEED = 102
raster_seed = BASE_SEED + 101

outdir = Path("../images"); outdir.mkdir(parents=True, exist_ok=True)
MAX_WORKERS = min(8, len(delta_grid))
```

## Sampling Mechanism

```python
def make_urn(size, p_one, rng):
    ones = int(np.clip(round(p_one * size), 0, size))
    urn = np.empty(size, dtype=np.int8)
    urn[: size - ones] = 0
    urn[size - ones :] = 1
    rng.shuffle(urn)
    return urn

def sample_without_replacement(urn, k, rng):
    idx = rng.choice(len(urn), size=k, replace=False)
    return urn[idx]
```

### CPP score

```python
def weighted_mean_score(x, t):
    
    x = np.asarray(x, dtype=float); n = x.shape[0]
    
    idx = np.arange(n, dtype=float)
    w = 1-np.abs(idx - t) / n
    w = np.clip(w, 0.0, None)  

    w1 = w[:t]; x1 = x[:t]; s1 = w1.sum()
    m_left = (x1 @ w1) / s1 if s1 > 0 else 0.0

    w2 = w[t:]; x2 = x[t:]; s2 = w2.sum()
    m_right = (x2 @ w2) / s2 if s2 > 0 else 0.0
        
    return float(abs(m_left - m_right))
```

## Compute confidence set for different $\delta$

```python
def confidence_set_urn_model(delta, seed, n=800, xi=350, nperm=500, alpha=0.05):
    rng = np.random.default_rng(seed)
    urn_size = 2500
    urn1 = make_urn(urn_size, 0.5 - delta, rng)
    urn2 = make_urn(urn_size, 0.5 + delta, rng)
    left  = sample_without_replacement(urn1, xi, rng)
    right = sample_without_replacement(urn2, n - xi, rng)
    x = np.concatenate([left, right])

    pseed = seed + 101
    pvals = compute_p_values(x, weighted_mean_score, nperm=nperm, random_state=pseed)

    return np.flatnonzero(pvals > alpha) + 1  
```

```python
def membership_from_indices(cs, n: int) -> np.ndarray:
    idx = np.asarray(cs, dtype=int).ravel()
    if idx.size == 0:
        return np.zeros(n - 1, dtype=float)
    v = np.zeros(n - 1, dtype=float)
    v[idx - 1] = 1.0
    return v


def stats_for_delta(delta: float, base_seed: int, R: int, nperm: int, alpha: float, n: int):
    sizes = np.empty(R, dtype=float)
    rows = []
    for r in range(R):
        idx_list = confidence_set_urn_model(delta, seed=base_seed + 10*r, nperm=nperm, alpha=alpha)
        sizes[r] = len(idx_list)
        rows.append(membership_from_indices(idx_list, n))
    memb = np.vstack(rows)
    mean_m = memb.mean(axis=0)
    se_m   = memb.std(axis=0, ddof=1)/np.sqrt(R) if R > 1 else np.zeros(n - 1)
    se_sz  = sizes.std(ddof=1)/np.sqrt(R) if R > 1 else 0.0
    return sizes.mean(), se_sz, mean_m, se_m

```

```python
# --- Allocate outputs ---
mean_cs = np.empty(delta_grid.size, dtype=float)
se_cs   = np.empty(delta_grid.size, dtype=float)
mean_membership = np.empty((delta_grid.size, n - 1), dtype=float)   # rows = deltas, cols = t
se_membership   = np.empty((delta_grid.size, n - 1), dtype=float)

# --- Parallel map over delta_grid ---
with ThreadPoolExecutor(max_workers=MAX_WORKERS) as ex:
    fut_map = {
        ex.submit(stats_for_delta, float(p), BASE_SEED + i, R, N_PERM, ALPHA, n): i
        for i, p in enumerate(delta_grid)
    }
    for fut in as_completed(fut_map):
        i = fut_map[fut]
        m_sz, se_sz, m_mem, se_mem = fut.result()
        mean_cs[i] = m_sz
        se_cs[i]   = se_sz
        mean_membership[i, :] = m_mem
        se_membership[i, :]   = se_mem

```

```python
# --- Plot: average membership raster + size curve ---
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Average inclusion probability Pr(t in CS) across R runs for each delta
im = ax1.imshow(
    mean_membership,
    aspect="auto",
    interpolation="nearest",
    origin="lower",
    extent=[1, n - 1, float(delta_grid[0]), float(delta_grid[-1])],
    vmin=0.0, vmax=1.0,
)
ax1.set_xlabel("t")
ax1.set_ylabel(r"$\delta$")
ax1.set_title("Average CS membership")
cbar = fig.colorbar(im, ax=ax1, fraction=0.046, pad=0.04)
cbar.set_label("Inclusion probability")

# CS length vs delta with 1 SE ribbon
ax2.plot(delta_grid, mean_cs, marker="o", linewidth=2.0, label="Mean size")
ax2.fill_between(delta_grid, mean_cs - se_cs, mean_cs + se_cs, alpha=0.25, label="1 SE")
ax2.set_title(r"Confidence set size vs. $\delta$")
ax2.set_xlabel(r"$\delta$")
ax2.set_ylabel("CS length")
ax2.legend()

fig.tight_layout()
outpath_avg = outdir / f"Confidence_set_urn_model.pdf"
fig.savefig(outpath_avg, dpi=200, bbox_inches="tight")
plt.show()
plt.close(fig)

print(f"Saved: {outpath_avg}")
```
